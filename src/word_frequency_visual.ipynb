```python
import matplotlib.pyplot as plt
import seaborn as sns


datasets = ['Tokyo Olympic Tweets', 'Apple Company Tweets', 'Neutral Politics Tweets']
top_words = [
    [('tokyo', 111229), ('olympics', 39097), ('gold', 15459), ('you', 13216), ('medal', 13213),
     ('i', 12860), ('olympic', 11907), ('first', 11898), ('womens', 8998), ('team', 8991)],
    [('sxsw', 9545), ('mention', 7054), ('link', 4285), ('rt', 2931), ('ipad', 2875),
     ('google', 2337), ('apple', 2146), ('iphone', 1519), ('store', 1468), ('new', 1089)],
    [('other', 476), ('some', 461), ('more', 460), ('trump', 460), ('president', 408),
     ('does', 396), ('us', 364), ('like', 318), ('government', 302), ('one', 299)]


unique_words = [185600, 10343, 29295]
elapsed_time_basic = [3.48, 0.09, 0.09]
elapsed_time_fast = [2.59, 0.08, 0.08]


def plot_top_words():
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    for i, dataset in enumerate(datasets):
        words, counts = zip(*top_words[i])
        axes[i].barh(words, counts, color='skyblue')
        axes[i].set_title(f"Top 10 Words in {dataset}")
        axes[i].set_xlabel("Word Frequency")
        axes[i].set_ylabel("Words")
    plt.tight_layout()
    plt.show()

def plot_results():
    fig, ax = plt.subplots(1, 2, figsize=(15, 5))

    ax[0].bar(datasets, unique_words, color='lightgreen')
    ax[0].set_title("Total Unique Words")
    ax[0].set_ylabel("Unique Words Count")

    bar_width = 0.35
    index = range(len(datasets))
    ax[1].bar(index, elapsed_time_basic, bar_width, label='Basic Version', color='lightcoral')
    ax[1].bar([i + bar_width for i in index], elapsed_time_fast, bar_width, label='Fast Version', color='skyblue')
    ax[1].set_title("Elapsed Time Comparison (Basic vs Fast Version)")
    ax[1].set_ylabel("Elapsed Time (seconds)")
    ax[1].set_xticks([i + bar_width / 2 for i in index])
    ax[1].set_xticklabels(datasets)
    ax[1].legend()

    plt.tight_layout()
    plt.show()


plot_top_words()
plot_results()
```
