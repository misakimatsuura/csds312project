```python
# Import dependencies
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Create VADER analyzer
analyzer = SentimentIntensityAnalyzer()

# Define a sentence to analyze, this will come from the social media data in the future
sentence = "How positive is this sentence?"
getScore = analyzer.polarity_scores

# Get the VADER analysis for the sentence and print the result
vs = getScore(sentence)
print("{:.<65} {}".format(sentence, str(vs)))
```
