```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from collections import defaultdict

```

```python
# Download VADER
nltk.download('vader_lexicon')

# Load NRC Emotion Lexicon
nrc_path = "https://raw.githubusercontent.com/ryanmcdermott/trump-speeches/master/data/NRC-Emotion-Lexicon-Wordlevel-v0.92.txt"


```

```python
# Load NRC Lexicon into a dictionary
nrc_emotions = defaultdict(set)
with nltk.request.urlopen(nrc_path) as file:
    for line in file:
        word, emotion, presence = line.strip().split("\t")
        if int(presence) == 1:
            nrc_emotions[emotion].add(word)

# Function to classify emotions in a sentence
def emotion_analysis(text):
    words = nltk.word_tokenize(text.lower())
    emotion_scores = {emotion: sum(1 for word in words if word in word_set) for emotion, word_set in nrc_emotions.items()}
    return max(emotion_scores, key=emotion_scores.get) if sum(emotion_scores.values()) > 0 else "Neutral"

# Example
text = "I am so grateful for all the love and support! I could not be happier"
print(emotion_analysis(text))  # Output: 'Joy' or 'Gratitude' (if gratitude is mapped)
```
